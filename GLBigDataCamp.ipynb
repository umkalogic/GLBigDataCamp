```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("test-task_dataset_summer_products.csv")
df.head()
```

```python
meanprice_all_products_foreach_country = df.groupby(['origin_country']).mean()
mean_all_products_foreach_country
```

```python
mean_products_by_country = df.groupby(['origin_country'], as_index=False)[['price']].mean().sort_values('origin_country')
mean_products_by_country
```

```python
mean_each_product_by_country = df.groupby(['origin_country','product_id'], as_index=False)[['price']].mean().sort_values('origin_country')
mean_each_product_by_country
```

```python
df1 = df.groupby(['origin_country'], as_index=False
                ).agg({'price': 'mean',
                       'rating_five_count': lambda x: x.sum(),
                       'rating_count':lambda x: x.sum()}
                     ).sort_values('origin_country').replace(0, np.nan)
#df1
```

```python
df2 = df1.dropna().assign(five_percentage=lambda x: ( x['rating_five_count'] / x['rating_count']) *100 )
df2.drop(['rating_five_count', 'rating_count'], axis='columns', inplace=True)
```

```python
df2  #SQL Representation select avg(price), (sum(rating_five_count) / sum(rating_count)) * 100 as five_percentage, origin_country
#from summer_products
#group by origin_country
#order by origin_country
```

```python
df3 = df.groupby(['origin_country','product_id'], as_index=False
                ).agg({'price': 'mean',
                       'rating_five_count':'sum',
                      'rating_count':'sum'}).sort_values('origin_country').replace(0, np.nan)
```

```python
df4 = df3.dropna().assign(five_percentage=lambda x: ( x['rating_five_count'] / x['rating_count']) *100 )
df4.drop(['rating_five_count', 'rating_count'], axis='columns', inplace=True)
```
